# 00. 환경 확인

논문: *Long-Term Modeling of Financial Machine Learning for Active Portfolio Management* (arXiv:2301.12346)

- 라이브러리 버전 / GPU 확인
- 공통 설정값 정의, seed 고정
- `data/`, `reports/` 폴더 생성

```python
# Cell 1: 라이브러리 import
import sys
import random
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import torch

try:
    import sklearn
    SKLEARN_VERSION = sklearn.__version__
except ImportError:
    SKLEARN_VERSION = "(not installed - 필수 아님)"
```

```python
# Cell 2: 버전 출력 및 GPU 확인
print("python     :", sys.version.split()[0])
print("numpy      :", np.__version__)
print("pandas     :", pd.__version__)
print("matplotlib :", matplotlib.__version__)
print("sklearn    :", SKLEARN_VERSION)
print("torch      :", torch.__version__)
print("cuda       :", torch.cuda.is_available())
if torch.cuda.is_available():
    print("gpu        :", torch.cuda.get_device_name(0))
```

```python
# Cell 3: 공통 설정값 정의 (seed 고정)
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("device:", DEVICE)
```

```python
# Cell 4: data / reports 폴더 생성
ROOT = Path("..").resolve()
DATA_DIR = ROOT / "data"
RAW_DIR = DATA_DIR / "raw"
REPORT_DIR = ROOT / "reports"
FIG_DIR = REPORT_DIR / "figures"
RESULT_DIR = REPORT_DIR / "results"

for path in [RAW_DIR, DATA_DIR / "processed", DATA_DIR / "sample", FIG_DIR, RESULT_DIR]:
    path.mkdir(parents=True, exist_ok=True)
    print("ok:", path.relative_to(ROOT))
```
